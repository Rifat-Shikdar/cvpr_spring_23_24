```python
set1 = {1, 3, 5, 9}
set2 = {1, 3, 7}
print(set1)
print(set2)
```

```python
unionSet = set1.union(set2)
#Uniion of both sets
print("Union product:", unionSet)
```

```python
intersectionSet = set1.intersection(set2)
#Intersection of 2 sets
print("Intersection product:", intersectionSet)
```
